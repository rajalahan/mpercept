# Matplotlib
Matplotlib is a plotting library for Python. Since *pictures speak louder than words*, let's see some plots rather than talking about it.

```python
import numpy as np
import matplotlib.pyplot as plt

# to display plots within the notebook
%matplotlib inline
```

```python
x = np.arange(0, 10)
y = np.arange(0, 10)
```

```python
plt.plot(x, y)
```

```python
# Plotting a line y=mx+c
x = np.arange(0, 20)
m = 1.5
c = 2
y = m*x + c

plt.plot(x, y)

plt.title("y=mx+c")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid()

plt.show()
```

```python
# Multiple plots in one
x = np.arange(0, 20)

m1, m2, m3 = 1.5, 4, 3
c1, c2, c3 = 2, 0, 2.5

y1 = m1*x + c1
y2 = m2*x + c2
y3 = m3*x + c3

plt.plot(x, y1, label='1st line')
plt.plot(x, y2, label='2nd line')
plt.plot(x, y3, label='3rd line')

plt.title('y=mx+c')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()

plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)

plt.title('The sine wave')
plt.xlabel('Angle')
plt.ylabel('Value')

plt.show()
```

```python
x = np.arange(0, 20)
y = x**2

plt.plot(x, y, color='red', linestyle='dashed')

plt.xlim(0, 10)
plt.ylim(0, 50)

plt.show()
```

```python
plt.plot(x, x, 'r--')
plt.plot(x, x**2, 'b*')
plt.plot(x, x**3, 'g-.')

plt.show()
```

```python
# Scatter plot
np.random.seed(0)
x = np.random.randint(0, 20, 40)
y = np.random.randint(0, 20, 40)
plt.scatter(x, y)
```

```python
plt.plot(x, y)
```

```python
np.random.seed(0)
x = np.random.randint(0, 100, 60)
y1 = np.random.randint(0, 100, 60)
y2 = np.random.randint(50, 150, 60)
y3 = np.random.randint(100, 200, 60)

plt.scatter(x, y1, label='Data 1', color='red')
plt.scatter(x, y2, label='Data 2', color='green')
plt.scatter(x, y3, label='Data 3', color='blue')

plt.legend()
plt.show()
```

```python
x = np.arange(0, 20)
y = np.random.randint(0, 30, 20)
plt.scatter(x, y)
plt.plot(x, 2*x, 'r')
plt.show()  # until show() is invoked, the plots will overlap
```

```python
np.random.seed(0)
x = np.random.randint(0, 100, 400)
y = np.random.randint(0, 100, 400)
color = ['red' if p>50 else 'blue' for p in y]
plt.scatter(x, y, color=color)

plt.show()
```

```python
# Multiple figures
x = np.linspace(0, 5, 100)
plt.subplot(2,2,1)  # grid size (2, 2), figure number 1
plt.plot(x, 2*x)

plt.subplot(2,2,2)
plt.plot(x, x**2)

plt.subplot(2,2,3)
plt.plot(x, x**0.5)

plt.subplot(2,2,4)
plt.plot(x, np.exp(x))

plt.show()
```

```python
# Bar plot
x = np.array([1, 2, 3, 4])
y = np.array([50, 30, 20, 10])
labels = ['Python', 'Java', 'C++', 'Others']
plt.bar(x, y, tick_label=labels)

plt.title('Popularity Survey')
plt.ylabel('Percentage')

plt.show()
```

```python
plt.pie(y, labels=labels, autopct='%.2f%%')
plt.show()
```
